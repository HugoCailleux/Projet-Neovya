### $\bullet$ Visualisation des capteurs sur une carte

```python
LATS = list(df_sensors['lat'])

LONGS = list(df_sensors['lon'])
```

Les coordonnées de la ville de Bordeaux sont les suivantes:

| $\text{Latitude de Bordeaux}$ | $\text{Longitude de Bordeaux}$ |
|-----|-----|
| $44.837789$ | $-0.57918$ |

```python
coords = (44.837789, -0.57918)
map = folium.Map(location=coords, tiles='OpenStreetMap', zoom_start=12)
```

```python
for i in range(len(df_sensors)):
    folium.CircleMarker(
        location = (LATS[i], LONGS[i]),
        radius = 1,
        color = 'crimson',
        popup= df_sensors.iloc[i]['sensor'],
        fill = True,
        fill_color = 'crimson'
    ).add_to(map)
```

```python
sensor_full = ['5143.01 2', '5146.12 2', '5146.15 1', '5044.12 2', 'LAPI L41', '5044.01 1', 
               '5742.00 1', '5146.13 1', '5143.00 2', '5044.15 1', '5042.00 1', '5043.00 1', 
               '5146.14 2', '5146.10 2', '5143.01 1', '5342.00 1', 'LAPI L42', '5146.11 1', 
               '5044.10 2', '5042.00 2', '5742.00 2', '5044.01 2', '5044.11 1', '5442.00 2']
```

```python
def df_concat(df,target_list):
    col_list = list(df.columns)
    df_target = pd.DataFrame(columns=col_list)
    for target in target_list:
        df1 = df.loc[df['sensor'].str.contains(target, regex=False)]
        df_target = pd.concat([df_target, df1],ignore_index=True)
    return df_target
```

```python
df_sensors_full = df_concat(df_sensors,sensor_full)
df_sensors_full
```

```python
lats = list(df_sensors_full['lat'])

longs = list(df_sensors_full['lon'])
```

```python
map_sensors = folium.Map(location=coords, tiles='OpenStreetMap', zoom_start=13)
```

```python
for i in range(len(df_sensors_full)):
    folium.CircleMarker(
        location = (lats[i], longs[i]),
        radius = 5,
        color = 'crimson',
        popup= df_sensors_full.iloc[i]['sensor'],
        fill = True,
        fill_color = 'crimson'
    ).add_to(map_sensors)
```
